```python
#plot 1
#line plot

# x-axis values 
x = [8,4,9,1,4] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot 
plt.plot(x,y) 
  
# function to show the plot 
plt.show()
```

```python
#barplot

# x-axis values 
x = ["car", "bike", "bus", "truck", "bicycle"] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot the bar 
plt.bar(x,y) 
  
# function to show the plot 
plt.show() 
```

```python
#creating a histogram plot

# Y-axis values 
y = [11,11,11, 15, 18, 2,3,3,3,1,3,4, 2] 
  
# Function to plot histogram 
plt.hist(y) 


plt.xlabel('quantity')
plt.ylabel('no. of its instances')
plt.title("basic histogram")

#Function to show the plot 
plt.show() 
```

```python
#trying to plot some maths

x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Mathematic graphs")

plt.legend()

plt.show()
```

```python
#Formatting the style of your plot

plt.plot([5, 10, 15, 20], [1, 8, 9, 16], 'ro')

plt.show()
```

```python
def y1(t):
    return ((3*t**2)-5*t) * np.sin(2*np.pi*t)

t1= np.arange(0.0, 10.0, 0.1)
t2= np.arange(0.0, 10.0, 0.05)

plt.figure(figsize=(20,12))
plt.subplot(211)
plt.plot(t1, y1(t1), 'bo', t2, y1(t2), 'k')
```

```python
plt.figure()  # create a plot figure

x1= np.linspace(0, 10, 100)
# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x1, np.sin(x1))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x1, np.cos(x1))
```

```python
#This is the best way to create plots. The idea here is to create Figure objects and call methods off it. 
```

```python
#scatter plot

x = np.random.randn(1, 100)
y = np.random.randn(1,100)
plt.scatter(x, y, color='blue', s=30) ## here s is size of point in scatter plot
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot Example')
plt.show()

```

```python
# ax will be an array of two Axes objects
fig, ax = plt.subplots(4)


x=np.arange(1,10,0.01)
# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x, 3*x**2)
ax[3].plot(x, (7*x**4+5*x**2-50))

```

```python

x=np.arange(-10,10,0.01)

plt.plot(x, np.sin(x))
plt.xlim(-5, 5)
plt.ylim(-5, 5);
```

```python
# assigning the legends

plt.plot(x, np.sin(x), 'blue', label='sin(x)')
plt.plot(x, np.cos(x), "red", label='cos(x)')
plt.axis('equal')
plt.legend();
```
